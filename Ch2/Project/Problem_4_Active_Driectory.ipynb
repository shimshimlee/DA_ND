```python
class Group(object):
    def __init__(self, _name):
        self.name = _name
        self.groups = []
        self.users = []

    def add_group(self, group):
        self.groups.append(group)

    def add_user(self, user):
        self.users.append(user)

    def get_groups(self):
        return self.groups

    def get_users(self):
        return self.users

    def get_name(self):
        return self.name




def is_user_in_group(user=None, group=Group("default")):
    """
    Return True if user is in the group, False otherwise.

    Args:
      user(str): user name/id
      group(class:Group): group to check user membership against
    """
    if (user == None) | ((group.get_groups()==[]) & (group.get_users==[])):
        return False
    
    users = group.get_users()
    if user in users :
        return True
    else :
        groups = group.get_groups()
        for each in groups :
            if is_user_in_group(user, each):
                return True
    
    return False

## Add your own test cases: include at least three test cases
## and two of them must include edge cases, such as null, empty or very large values

## Test Case 1
print ("Test Case 1 : Udacity Test Example")
parent = Group("parent")
child = Group("child")
sub_child = Group("subchild")

sub_child_user = "sub_child_user"
sub_child.add_user(sub_child_user)

child.add_group(sub_child)
parent.add_group(child)

print(parent.get_name(), "group has a sub_child_user?",
      is_user_in_group(sub_child_user, parent))
print(child.get_name(), "group has a sub_child_user?",
      is_user_in_group(sub_child_user, child))
print(sub_child.get_name(), "group has a sub_child_user?",
      is_user_in_group(sub_child_user, sub_child))

## Test Case 2
print ("Test Case 2 : Edge case - empty group")
empty_group = Group("empty")
print(empty_group.get_name(), "group has any user?", is_user_in_group("any",empty_group))

## Test Case 3
print("Test Case 3 : Edge case - deep sub group and user")
head = Group("head")
sub1 = Group("sub1")
sub2 = Group("sub2")
sub3 = Group("sub3")
deep_user = "deep_user"

head.add_group(sub1)
sub1.add_group(sub2)
sub2.add_group(sub3)
sub3.add_user(deep_user)
print(head.get_name(), "group has a deep_user?",
      is_user_in_group(deep_user, head))
```
