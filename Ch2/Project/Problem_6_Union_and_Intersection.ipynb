```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

    def __repr__(self):
        return str(self.value)


class LinkedList:
    def __init__(self):
        self.head = None

    def __str__(self):
        cur_head = self.head
        out_string = ""
        while cur_head:
            out_string += str(cur_head.value) + " -> "
            cur_head = cur_head.next
        return out_string


    def append(self, value):

        if self.head is None:
            self.head = Node(value)
            return

        node = self.head
        while node.next:
            node = node.next

        node.next = Node(value)

    def size(self):
        size = 0
        node = self.head
        while node:
            size += 1
            node = node.next

        return size

def union(llist_1, llist_2):
    # Your Solution Here
    union = LinkedList()
    set1 = set()
    set2 = set()
    
    if llist_1.head is None and llist_2.head is None:
        return None
    
    cur1_head = llist_1.head
    while cur1_head is not None :
        set1.add(cur1_head.value)
        cur1_head = cur1_head.next
    
    cur2_head = llist_2.head
    while cur2_head is not None :
        set2.add(cur2_head.value)
        cur2_head = cur2_head.next
    
    set12 = set1.union(set2)
    
    for i in set12 :
        union.append(i)
    
    return union

def intersection(llist_1, llist_2):
    intersection = LinkedList()
    set1 = set()
    set2 = set()
    
    if llist_1.head is None and llist_2.head is None:
        return None
    
    cur1_head = llist_1.head
    while cur1_head is not None :
        set1.add(cur1_head.value)
        cur1_head = cur1_head.next
    
    cur2_head = llist_2.head
    while cur2_head is not None :
        set2.add(cur2_head.value)
        cur2_head = cur2_head.next
    
    set12 = set1.intersection(set2)
    
    if len(set12) == 0:
        return None
    
    for i in set12:
        intersection.append(i)
    
    return intersection


## Test case 1

linked_list_1 = LinkedList()
linked_list_2 = LinkedList()

element_1 = [3,2,4,35,6,65,6,4,3,21]
element_2 = [6,32,4,9,6,1,11,21,1]

for i in element_1:
    linked_list_1.append(i)

for i in element_2:
    linked_list_2.append(i)

print (union(linked_list_1,linked_list_2))
print (intersection(linked_list_1,linked_list_2))

## Test case 2 (in case of that Intersection is null)

linked_list_3 = LinkedList()
linked_list_4 = LinkedList()

element_1 = [3,2,4,35,6,65,6,4,3,23]
element_2 = [1,7,8,9,11,21,1]

for i in element_1:
    linked_list_3.append(i)

for i in element_2:
    linked_list_4.append(i)

print (union(linked_list_3,linked_list_4))
print (intersection(linked_list_3,linked_list_4))

## Test case 3 (in case of that set1 is null)
linked_list_5 = LinkedList()
linked_list_6 = LinkedList()

element_1 = []
element_2 = [1,7,8,9,11,21,1]

for i in element_1:
    linked_list_5.append(i)

for i in element_2:
    linked_list_6.append(i)
    
print (union(linked_list_5,linked_list_6))
print (intersection(linked_list_5,linked_list_6))
```
